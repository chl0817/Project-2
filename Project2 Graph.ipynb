```python
import matplotlib.pyplot as plt

# Data from 3.3
n_values = [1000, 100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, 900000, 1000000]
execution_times = [0.2611, 23.9450, 47.0692, 70.6414, 93.3721, 116.2842, 140.9817, 212.6533, 190.2085, 225.9621, 247.2601]
normalized_times = [ (t / n) * 1000 for t, n in zip(execution_times, n_values) ]  # Time per element in s

# Plotting the execution times
plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times, marker='o')
plt.title('Execution Time vs. Input Size for QuickSelect')
plt.xlabel('Input Size (n)')
plt.ylabel('Execution Time (ms)')
plt.grid(True)
plt.show()

# Plotting the normalized times
plt.figure(figsize=(10, 6))
plt.plot(n_values, normalized_times, marker='o')
plt.title('Normalized Execution Time per Element vs. Input Size for QuickSelect')
plt.xlabel('Input Size (n)')
plt.ylabel('Time per Element (s)')
plt.grid(True)
plt.show()
```
